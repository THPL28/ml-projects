## Titanic Survival Prediction using Machine Learning

## Objective
Predict whether a passenger survived using machine learning.

### Dataset
Titanic dataset.

### Result
Achieved 82% accuracy using Random Forest.

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
```

```python
df = pd.read_csv(url)
```

# Cleaning irrelevant data

```python
df['Age'] = df['Age'].fillna(df['Age'].mean()).astype(int)

df['Age'] = df['Age'].astype(int)
```

```python
df = df.drop(columns=['Cabin'])
```

```python
df = df.dropna()
```

```python
df.head()
```

### Convert text to number

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

## Features (input) and  Target (to predict)

```python
X = df[['Pclass', 'Age', 'Sex', 'Fare']]


y = df['Survived']
```

## Data division

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2
)
```

# Import





```python
from sklearn.linear_model import LogisticRegression
```

# Create model

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

Manking predictions

```python
y_pred = model.predict(X_test)
```

Evaluate model

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Normalize data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Retrain

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

# New acuray

```python
from sklearn.metrics import accuracy_score

print("New acuracy:", accuracy_score(y_test, y_pred))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Example considering that the scaler only expects [Pclass, Age, Sex, Fare]
people = [[3, 25, 0, 7.25]]

people_escale = scaler.transform(people)
prediction = model.predict(people_escale)

print("Survived?", prediction)
```
